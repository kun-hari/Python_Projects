```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load data
crime_df = pd.read_csv('crime_safety_dataset.csv')
crime_df.columns = [c.strip().lower() for c in crime_df.columns]
crime_df
```

```python
# ---------------- Core metrics ----------------
# 1. City / State with most crimes (assuming each row is a crime incident)
city_counts = crime_df['city'].value_counts()
state_counts = crime_df['state'].value_counts()

# 2. Most common crime type
type_counts = crime_df['crime_type'].value_counts()

# 3. Average victim age over all crimes
avg_age = crime_df['victim_age'].mean()

print('Top 5 cities by incidents:')
print(city_counts.head())
print('Top 5 states by incidents:')
print(state_counts.head())
print('Most common crime types:')
print(type_counts.head())
print('Average victim age overall:', avg_age)
```

```python
# 3. Visuals
plt.figure(figsize=(8,5))
city_counts.head(10).plot(kind='bar')
plt.title('Top 10 Cities by Number of Crime Incidents')
plt.ylabel('Number of Incidents'); plt.xlabel('City'); plt.tight_layout(); plt.show()

plt.figure(figsize=(8,5))
type_counts.head(10).plot(kind='bar', color='coral')
plt.title('Most Common Crime Types')
plt.ylabel('Number of Incidents'); plt.xlabel('Crime Type'); plt.tight_layout(); plt.show()

plt.figure(figsize=(8,5))
sns.histplot(crime_df['victim_age'], bins=15, kde=True)
plt.title('Victim Age Distribution')
plt.xlabel('Age'); plt.ylabel('Frequency'); plt.tight_layout(); plt.show()
```
