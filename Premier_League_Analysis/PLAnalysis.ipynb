```python
import pandas as pd
import numpy as np


data = pd.read_csv('C:\Hari\Work\Data analysis projects\Premier league\PremierLeagueMatches.csv')
data1 = data.iloc[:, :-1]
data1
```

```python
data1.drop_duplicates
data1.dropna (inplace = True) 
data1.rename(columns = {'Home Team': 'Home_team', 'Away Team':'Away_team'}, inplace=True)
data1
```

```python
#Total Home and away goals per team

total_home_goals_by_team = data1.groupby('Home_team')['homeScore'].sum()
print(total_home_goals_by_team.head(10))

total_away_goals_by_team = data1.groupby('Away_team')['awayScore'].sum()
total_away_goals_by_team.head(10)
```
